```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from scipy.interpolate import UnivariateSpline
```

```python
z = [0,1,1.4,2,2.5,3,4,5]
zz = [0,2,1.0,4,2.5,2,1,4]
```

```python
x = []
for item in z:
    x.append(5*item)
x
```

```python
y = []
for item in zz:
    y.append(4*item)
y
```

```python
interp_func = UnivariateSpline(x,y)
```

```python
newx = np.arange(0, 25, 0.1)
newy = interp_func(newx)
```

```python
plt.plot(newx, newy)
plt.show()
```

```python
yhat = newy + np.random.randn(len(newx))
```

```python
plt.plot(newx, yhat)
plt.show()
```

```python
newx = np.round_(newx,2)
newy = np.round_(newy,2)
```

```python
List = list(zip(newx,newy))
```

```python
string = ''
for item in List:
    string += str(item) + " -- "
```
